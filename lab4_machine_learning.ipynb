```python
from sklearn.datasets import load_iris
from sklearn.inspection import DecisionBoundaryDisplay
from matplotlib.colors import ListedColormap

import numpy as np
import matplotlib.pyplot as plot
import pickle

dataset = load_iris()
list(dataset.target_names)
colors = ListedColormap(["black", "blue", "red"]) 
```

```python
X, y = dataset.data[:, :2],dataset.target
plot.scatter(X[:,0], X[:,1], c=y, cmap=colores)
plot.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3) 

plot.scatter(x_train[:,0],x_train[:,1], c=y_train, cmap=colors)
plot.scatter(x_test[:,0],x_test[:,1], c=y_test, alpha=0.2, cmap=colors)
plot.xlabel("sepal length (cm)")
plot.ylabel("sepal width (cm)")
plot.show()
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(x_train, y_train)

print("train data: {}".format(model.score(x_train, y_train)))
print("test data: {}".format(model.score(x_test, y_test)))

graph = DecisionBoundaryDisplay.from_estimator(model, X, cmap=colors, alpha=0.3)
graph.ax_.scatter(x_train[:,0],x_train[:,1], c=y_train, cmap=colors)
plot.xlabel("sepal length (cm)")
plot.ylabel("sepal width (cm)")
plot.show()
```

```python
def eval_model(model, x, y):
    
    x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
    model.fit(x_train, y_train)

    print("train data: {}".format(model.score(x_train, y_train)))
    print("test data: {}".format(model.score(x_test, y_test)))

    graph = DecisionBoundaryDisplay.from_estimator(model, X, cmap=colors, alpha=0.4)
    graph.ax_.scatter(x_train[:,0],x_train[:,1], c=y_train, cmap=colors)
    plot.xlabel("sepal length (cm)")
    plot.ylabel("sepal width (cm)")
    plot.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(5)
eval_model(model, x, y)
pickle.dump(model, open("./modelo_entrenado.sav", 'wb'))
```
